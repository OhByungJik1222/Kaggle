# **Result**

## **Linear Regression**

```python
lr_result = lr_result.sort_values(by='MAE').head()
```

```python
lr_result.reset_index(drop=True).drop(columns='Unnamed: 0')
```

```python
g = sns.catplot(
    data=lr_result,
    x='Model',
    y='MAE',
    hue='Scaler',
    col='Data',
    kind='bar',
    height=5,
    aspect=1.2
)

g.fig.suptitle("Model Performance (MAE) by Scaler & Data", y=1.05)
plt.show()
```
